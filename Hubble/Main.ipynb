```python
import matplotlib.pyplot as plt
from astropy.io import fits

# Open the FITS file
hdul = fits.open("f001a066.fits")

# Show information about the HDUs
hdul.info()

# Extract the image data (assuming its in the first HDU)
image_data = hdul[0].data

# Display the image
plt.imshow(image_data, cmap='gray')
plt.colorbar()
plt.show()

# Close the FITS file
hdul.close()
```

```python
import matplotlib.pyplot as plt
from astropy.io import fits

# Open the FITS file
hdul = fits.open("f001a066.fits")
image_data = hdul[0].data
hdul.close()

# Create a false-color image using a colormap, e.g., 'plasma' or 'inferno'
plt.imshow(image_data, cmap='plasma')
plt.colorbar()  # Optional: Adds color bar to show intensity values
plt.title("False-Color Image (Plasma Colormap)")
plt.show()
```

```python
import matplotlib.pyplot as plt
from astropy.io import fits

# Open the FITS file
hdul = fits.open("f001a066.fits")
image_data = hdul[0].data
hdul.close()

# Apply a false color map, e.g., 'plasma' or 'inferno'
plt.imshow(image_data, cmap='inferno')
plt.colorbar()  # Optional: Add color bar to represent intensity values
plt.title("False-Color Image (Inferno Colormap)")
plt.show()
```

```python
from matplotlib.colors import LinearSegmentedColormap

# Define a custom colormap with specific color stops
custom_cmap = LinearSegmentedColormap.from_list("custom", [ "blue", "green", "red"])

# Apply the custom colormap to an image
plt.imshow(image_data, cmap=custom_cmap)
plt.colorbar()
plt.title("Custom False-Color Image")
plt.show()
```
